```r
#change headings to what I want and to match other datasets
policy_data <- policy_data %>%
    mutate(
        State = State.Territory,
        Policy_Type = Policy.Incentive.Type,
        Year = Created
    ) %>%
    
    #want year only
    mutate(Created = as.Date(Created, format = "%m/%d/%Y")) %>%
    mutate(Year = format(Created, "%Y")) %>% #extract year from full date

    select(State, Year, Policy_Type)

#policy_data
```

```r
#create New_Policies, Cumulative_Policies and Past_Policies variables

yearly_policy_data <- policy_data %>%
    select(State, Year) %>%
    group_by(State, Year) %>% #counts number of policy rows by group_by
        mutate(New_Policies = n()) %>%
        distinct() %>%

  #want to make cumulative yearly column
  #you must first ungroup() if you want the overall cumulative sum across all groups
    ungroup() %>% #MUST ungroup by above
    mutate(Cumulative_Policies = cumsum(New_Policies)) %>% #cumulative yearly policies
    mutate(Past_Policies = Cumulative_Policies - New_Policies) %>%

    select(State, Year, New_Policies, Cumulative_Policies)

#yearly_policy_data
```
