```r
#load in packages
library(tidyverse)

#load in raw data
policy_data <- read.csv(file.path(file_path, "raw_data/State_Clean_Energy_Policies.csv"))
```

```r
#look at data
head(policy_data)

#Want State.Terrtory, Policy.Incentive.Type and Created
    #State.Terrtory -> State
    #Policy.Incentive.Type -> Policy_Type
    #Created -> Year (Want year only)
```

```r
#change headings to what I want and to match other datasets
policy_data <- policy_data %>%
    mutate(
        State = State.Territory,
        Policy_Type = Policy.Incentive.Type,
        Year = Created
    ) %>%
    
    #want year only
    mutate(Created = as.Date(Created, format = "%m/%d/%Y")) %>%
    mutate(Year = format(Created, "%Y")) %>% #extract year from full date
    select(State, Year, Name)

policy_data #missing years
```

```r
#create New_Policies, Cumulative_Policies and Past_Policies variables
yearly_policy_data <- policy_data %>%
    select(State, Year) %>%
    group_by(State, Year) %>% #counts number of policy rows by group_by
        mutate(New_Policies = n()) %>%
        distinct()

yearly_policy_data
#do not make cumulative yearly column - have to redo in combined state dataset anyways
```

```r
#checks for NAs
yearly_policy_data_NAs <- yearly_policy_data[!complete.cases(yearly_policy_data), ]
yearly_policy_data_NAs
```

```r
write.csv(yearly_policy_data, file.path(file_path, "Lesson_1:Cleaning_Data_for_Analysis/cleaned_datasets/State_Yearly_Policies.csv"))
```
