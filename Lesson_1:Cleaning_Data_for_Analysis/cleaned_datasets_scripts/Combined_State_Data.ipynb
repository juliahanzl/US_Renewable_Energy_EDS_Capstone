```r
#load in packages
library(tidyverse)

#load in raw data
political_party <- read.csv(file.path(file_path, "cleaned_datasets/State_Political_Party.csv"))
energy_consumption <- read.csv(file.path(file_path, "cleaned_datasets/State_Energy_Consumptions.csv"))
yearly_policies <- read.csv(file.path(file_path, "cleaned_datasets/State_Yearly_Policies.csv"))
```

```r
#look at data
head(political_party)
head(energy_consumption)
head(yearly_policies)
```

```r
all_energy <- political_party %>%
    inner_join(energy_consumption, by = c("State", "Year")) %>%
    select(State, Year, Political_Party, Total_Energy, NonRenewable_Energy, Perc_NonRenewable_Energy, Renewable_Energy, Perc_Renewable_Energy) %>%
    distinct()

all_energy

#checks for NAs
#all_data_NAs <- all_data[!complete.cases(all_data), ]
#all_data_NAs
```

```r
write.csv(all_energy, file.path(file_path, "/cleaned_datasets/State_PolParty_Energy.csv"))
```

```r
all_YearlyPolicies <- political_party %>%
    inner_join(yearly_policies, by = c("State", "Year")) %>%
    select(State, Year, Political_Party, New_Policies, Cumulative_Policies) %>%
    distinct()

all_YearlyPolicies
```

```r
write.csv(all_YearlyPolicies, file.path(file_path, "/cleaned_datasets/State_PolParty_YearlyPol.csv"))
```
