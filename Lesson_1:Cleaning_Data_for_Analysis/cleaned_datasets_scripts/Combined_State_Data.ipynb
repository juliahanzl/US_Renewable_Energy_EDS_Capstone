```r
#load in packages
library(tidyverse)

#load in raw data
political_party <- read.csv(file.path(file_path, "cleaned_datasets/State_Political_Party.csv"))
energy_consumption <- read.csv(file.path(file_path, "cleaned_datasets/State_Energy_Consumptions.csv"))
yearly_policies <- read.csv(file.path(file_path, "cleaned_datasets/State_Yearly_Policies.csv"))

political_party
#head(energy_consumption)
#head(yearly_policies)
```

```r
#look at data
#head(political_party)
political_party_NAs <- political_party[!complete.cases(political_party), ]
#political_party_NAs

#head(energy_consumption) 
energy_consumption_NAs <- energy_consumption[!complete.cases(energy_consumption), ]
#energy_consumption_NAs 

#head(yearly_policies) #missing years
yearly_policies_NAs <- energy_consumption[!complete.cases(energy_consumption), ]
#yearly_policies_NAs 
```

```r
#merge political part dataset with energy dataset
all_energy <- political_party %>%
    full_join(energy_consumption, by = c("State", "Year")) %>%
    select(State, Year, Political_Party, Total_Energy, NonRenewable_Energy, NonRenewable_Percent, Renewable_Energy, Renewable_Percent) %>%
    distinct() %>%
   drop_na()
    #filter(Year >= 2000 & Year <= 2023)

all_energy
```

```r
#checks for NAs
all_energy_NAs <- all_energy[!complete.cases(all_energy), ]
all_energy_NAs #87 NAs

#missing data for political party

#all_energy2 <- all_energy2 %>% 
#    filter(!(State != "US" & is.na(Political_Party))) #deletes all state Political_Party rows = NA unless US

#all_energy2
```

```r
#merge political party dataset with politcy dataset
yearly_policies <- political_party %>%
full_join(yearly_policies, by = c("State", "Year")) %>%
    select(State, Year, Political_Party, New_Policies) %>%
    filter(Year >= 2000 & Year <= 2023) %>%
    mutate(New_Policies = replace_na(New_Policies, 0)) %>%
    distinct()
   
yearly_policies

#missing Political Party and policies
```

```r
all_combined_data <- yearly_policies %>%
    full_join(all_energy, by = c("State", "Year", "Political_Party")) %>%
    drop_na()

all_combined_data
```

```r
write.csv(all_combined_data, file.path(file_path, "/cleaned_datasets/Combined_Data.csv"))
```
